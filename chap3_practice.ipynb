```python
import numpy as np
import plotly.graph_objects as go
```

```python
#exercise 3-1
def lwc(v, s):
    scal_list = list(s)
    vectors = np.array(v)
    Temp = np.zeros(3)
    for i in scal_list:
        Temp += vectors[scal_list.index(i)] * i
    return Temp
```

```python
lwc([[4,5,1],[-4,0,-4],[1,3,2]],[1,2,-3])
```

```python
l1 = 1
l2 = 2
l3 = -3
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])
l1*v1 + l2*v2 + l3*v3
```

```python
#exercise 3-3
def rpss(v):
    v_set = np.array(v)
    s = np.random.random_integers(low=-4,high=4, size=100)
    sub = []
    for i in v_set:
        for j in s:
           sub.append(i*j) 
    return np.array(sub)
```

```python
rp_sub = rpss([[1,3]])
rp_sub[:10]
```

```python
rp_sub2 = rpss([[3, 5, 1],[0, 2, 2]])
rp_sub2[:10]
```

```python
fig = go.Figure( data=[go.Scatter3d(
x=rp_sub2[:,0], y=rp_sub2[:,1], z=rp_sub2[:,2],
mode='markers' )])
fig.show()
```
