Dot Product
dot product is sum of corresponding elements multiplication. in other words, multiplying corresponding elements of 2 vectors and adding the resulting values together. It is used for mapping and measuring the similarity between two vectors.to use dot notation you simply call dot() method from Numpy.
np.dot()
the output will be a scalar number. We also can multiply vectors by an scalar before dot product.

```python
np.dot(a, b) , np.dot(3*a, b) #a multiplied by 3
```

dot product is distributive, just like the equations in basic math: a(b+c)= ab + ac
dot notation is possible only on the vectors of the same dimension.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([6,5,6])

num1 = np.dot(a, b+c)
num2 = np.dot(a,b) + np.dot(a,c)
num1 , num2,len(a), len(b), len(c)
```

as you notice the values of num1 and num2 are equal.

Broadcasting
Broadcasting is a technique in algebra that demonstrate the idea of taking one element of a matrix and multiplying it to every other element of the other matrix. in numpy to access broadcasting we use np.outer() function. as you have guesed, size of neither matrices matter.

```python
a = np.random.randn(2,2)
b = np.random.randn(3,3)
np.outer(b,a) #size of the product will be len(b)xlen(a)
```

K-means Clustering
is a method of clustering various multivarient data into a small group vased on the minimizing group center. It is basically choosing a number of centroids and computing the distance between our data and that centroid and assigning it to the group.

Algorithm Steps:
1. initialize centroids
2. compute Euclidean distance between the observations and centroids
3. assign each data to the group with the closest centroid
4. update each centroid as the average of all data assigned to it
5. repeat 2-4
