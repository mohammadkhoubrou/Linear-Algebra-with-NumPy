```python
import numpy as np
```

In order to use numpy library we need to add to our code. Usually we add it as np. short for numpy.

```python
python_list = [1,2,3]
np_array = np.array([1,2,3]) #1D array
```

We choose to use numpy arrays instead of python list to display and hold our data. With numpy arrays can represent vectors better and we can take some actions on arrays with the delighting result, that the python lists can't represent.(we'll see more example on this subject)

```python
v = np.array([[1,2,3]]) #row vector
w = np.array([[3],[4],[5]]) # column vector
```

Here we specify our vectors. There's a set of brackets to group all the outcome together and each aditional [] indicates a row.

```python
print(np_array.shape)
print(v.shape)
print(w.shape)
```

shape(): this function displays the orientation of the array. for 1D arrays we will only see one number in the parantesses that shows the number of elements in it. for row vectors and column vectors, we'll see number of row and colums.
