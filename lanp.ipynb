Dot Product
dot product is sum of corresponding elements multiplication. in other words, multiplying corresponding elements of 2 vectors and adding the resulting values together. It is used for mapping and measuring the similarity between two vectors.to use dot notation you simply call dot() method from Numpy.
np.dot()
the output will be a scalar number. We also can multiply vectors by an scalar before dot product.

```python
np.dot(a, b) , np.dot(3*a, b) #a multiplied by 3
```

dot product is distributive, just like the equations in basic math: a(b+c)= ab + ac
dot notation is possible only on the vectors of the same dimension.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([6,5,6])

num1 = np.dot(a, b+c)
num2 = np.dot(a,b) + np.dot(a,c)
num1 , num2,len(a), len(b), len(c)
```

as you notice the values of num1 and num2 are equal.
