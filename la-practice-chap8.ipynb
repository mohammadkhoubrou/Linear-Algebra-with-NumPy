```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
imgP = 20
mat = np.random.randn(imgP,imgP)
```

```python
k_size = 7
hk_size = k_size//2
x,y = np.meshgrid(np.linspace(-3,3,k_size),np.linspace(-3,3,k_size))
kernel = np.exp(-(x**2+y**2)/imgP)
kernel = kernel / np.sum(kernel)
```

```python
output = np.zeros((imgP+k_size-1,imgP+k_size-1))
imagePad = np.zeros(output.shape)
imagePad[hk_size:-hk_size:1,hk_size:-hk_size:1] = mat
for row in range(hk_size, imgP+hk_size):
    for col in range(hk_size, imgP+hk_size):
        image = imagePad[row - hk_size: row + hk_size + 1:1,
                    col - hk_size: col+ hk_size +1 :1]
        dot_p = np.sum(image * kernel)
        output[row, col] = dot_p
```

```python

fig,ax = plt.subplots(2,2,figsize=(8,8))

ax[0,0].imshow(image)
ax[0,0].set_title('Image')

ax[0,1].imshow(kernel)
ax[0,1].set_title('Convolution kernel')

ax[1,0].imshow(output)
ax[1,0].set_title('Manual convolution')


# for i in ax.flatten(): i.axis('off')

plt.savefig('Figure_07_04b.png',dpi=300)
plt.show()
```
