```python
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
#exercise 4-1
def pearson_cosine(v1,v2):
    vs = np.array([v1,v2])
    p = []
    for i in vs:
        mean = np.mean(i)
        p.append([j-mean for j in i])
    pearson = (np.dot(p[0],p[1]/np.linalg.norm(p[0])*np.linalg.norm(p[1])))
    cosine = np.dot(vs[0],vs[1]) / (np.linalg.norm(vs[0]) * np.linalg.norm(vs[1]))
    return pearson, cosine
```

```python
a =np.array([1,2,3,4])
b = np.array([5,6,7,8])
pearson_cosine(a,b)
```

```python
#exercise 4-2
var1 = [0,1,2,3]
var2 =[0.2, 1.2, 2.2, 3.2]
# cos = [np.dot(vs[0],vs[1]) / (np.linalg.norm(vs[0]) * np.linalg.norm(vs[1])) for i in range(-50,50) ]
# res = [pearson_cosine(var1, [round(j-i) for j in var2]) for i in range(-50,50)]
res = [np.array(pearson_cosine(var1, [round(j-i) for j in var2])) for i in range(-50,50)]
cos = [i[1] for i in res ]
pearson = [i[0] for i in res]
```

```python
x = range(-50, 50)

# Create the plot
sns.lineplot(x=x, y=cos, label='Cosine Similarity')
sns.lineplot(x=x, y=pearson, label='Pearson Correlation')
```
