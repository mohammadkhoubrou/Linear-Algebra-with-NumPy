```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.linalg import norm
from numpy import dot
import time
```

```python
#exercise 4-1
def pearson_cosine(v1,v2):
    vs = np.array([v1,v2])
    p = []
    for i in vs:
        mean = np.mean(i)
        p.append([j-mean for j in i])
    pearson = np.dot(p[0],p[1]) / (np.linalg.norm(p[0]) * np.linalg.norm(p[1]))
    cosine = np.dot(vs[0],vs[1]) / (np.linalg.norm(vs[0]) * np.linalg.norm(vs[1]))
    return pearson, cosine
```

```python
a =np.array([1,2,3,4])
b = np.array([5,6,7,8])
pearson_cosine(a,b)
```

```python
#exercise 4-4
def pearson(v1,v2):
    vs = [v1,v2]
    p = []
    for i in vs:
        mean = np.mean(i)
        p.append([j-mean for j in i])
    return dot(p[0],p[1]) / (norm(p[0]) * norm(p[1]))
```

```python
for i in range(1000):
    a = np.random.randint(10, size=(500))
    b = np.random.randint(10, size=(500))
    T_P_S = time.time()
    pearson(a,b)
    T_P_E = time.time()
    T_C_S = time.time()
    np.corrcoef(a,b)
    T_C_E = time.time()
print(f"pearson: {T_P_E - T_P_S} \ncorrcoef: {T_C_E - T_C_S}")
print((T_P_E - T_P_S)/(T_C_E - T_C_S))
```

```python
#exercise 4-6
kernel = np.array([0, .1, .3, .8, 1, .8,.3, .1, 0])
kernel = kernel/np.sum(kernel)
sns.lineplot(data= kernel)
np.sum(kernel)
```

```python
signal1 = np.random.randn(100)
sns.lineplot(data=signal1)
```

```python
edge1 = np.zeros(len(signal1))
```

```python
len_kernel = len(kernel)
kernel_half = len_kernel//2
len_signal =  len(signal1)
```

```python
for i in range(kernel_half+1, len_signal-kernel_half ):
    edge1[i] = np.dot(kernel, signal1[i-kernel_half-1: i + kernel_half])
edge1
```

We basically iterate the entire kernel on the whole signal, one sequence at a time. In order to place the entire kernel on the signal and calculate the dot product we have to watch out the slicing, so we can start from the begining of the signal and be able to use the kernel.

```python
sns.lineplot(data=edge1, label="fitered signal")
sns.lineplot(data=signal1, label="signal")
sns.lineplot(data=kernel, label="kernel")
```

```python
#exercise 4-7
def k_means(vectors, centroids, k):
    clusters = [[] for i in range(k)]
    for i in vectors:
        distance = [np.linalg.norm(i - j) for j in centroids]
        clusters[np.argmin(distance)].append(min(distance))
    
    new_centroids = np.array(np.mean(cluster, axis=0) for cluster in clusters)
    
    if np.all(new_centroids == centroids):
        pass
    
    centroids == new_centroids
    return centroids, clusters
```

```python
vectors = np.random.randn(150,2)
k = 2
```

```python
centroids = vectors[np.random.choice(vectors.shape[0], k, replace=False)]
centroid, cluster = k_means(vectors,centroids, k)
plt.scatter(x=vectors[:,0],y=vectors[:,1])
plt.scatter(x=centroid[:,0],y=centroid[:,1])
# vectors
```
