```python
import numpy as np
import matplotlib as plt
import seaborn as sns
from numpy.linalg import norm
from numpy import dot
import time
```

```python
#exercise 4-1
def pearson_cosine(v1,v2):
    vs = np.array([v1,v2])
    p = []
    for i in vs:
        mean = np.mean(i)
        p.append([j-mean for j in i])
    pearson = np.dot(p[0],p[1]) / (np.linalg.norm(p[0]) * np.linalg.norm(p[1]))
    cosine = np.dot(vs[0],vs[1]) / (np.linalg.norm(vs[0]) * np.linalg.norm(vs[1]))
    return pearson, cosine
```

```python
a =np.array([1,2,3,4])
b = np.array([5,6,7,8])
pearson_cosine(a,b)
```

```python
x = range(-50, 50)

# Create the plot
sns.lineplot(x=x, y=cos, label='Cosine Similarity')
sns.lineplot(x=x, y=pearson, label='Pearson Correlation')
```

```python
#exercise 4-4
def pearson(v1,v2):
    vs = [v1,v2]
    p = []
    for i in vs:
        mean = np.mean(i)
        p.append([j-mean for j in i])
    return dot(p[0],p[1]) / (norm(p[0]) * norm(p[1]))
```

```python
for i in range(1000):
    a = np.random.randint(10, size=(500))
    b = np.random.randint(10, size=(500))
    T_P_S = time.time()
    pearson(a,b)
    T_P_E = time.time()
    T_C_S = time.time()
    np.corrcoef(a,b)
    T_C_E = time.time()
print(f"pearson: {T_P_E - T_P_S} \ncorrcoef: {T_C_E - T_C_S}")
print((T_P_E - T_P_S)/(T_C_E - T_C_S))
```
