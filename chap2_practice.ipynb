```python
import numpy as np
import math
```

```python
def magnitude(Vector):
    Temp = 0
    for i in Vector:
        Temp += i**2
    return math.sqrt(Temp)
```

```python
magnitude([1,2,3,7,8,9])
```

```python
np.linalg.norm([1,2,3,7,8,9])
```

```python
# def unit_v(Vector):
#     Temp = 1/magnitude(Vector)
#     return Temp*np.array(Vector)
def unit_v(Vector):
    Temp = 1/np.linalg.norm(Vector)
    return Temp*np.array(Vector)
```

```python
unit_v([0,0,0,0,0,0])
```

```python
unit_v([1,2,3,7,8,9])
```

```python
def convertor(Row):
    column =[]
    for j in Row:
        for i in j:
            i = [i]
            column.append(i)
    return column
```

```python
convertor([[1,2,3,7,8,9]])
```

```python
def unit_vector(vector, mag):
    unit = unit_v(vector)
    return mag * unit
```

```python
unit_vector([1,2,3,7,8,9],2)
```

```python
def sqr_norm(vector):
    Temp = 0
    for i in vector:
        Temp += i**2
    return Temp
```

```python
print(magnitude([1,2,3,7,8,9])**2)
sqr_norm([1,2,3,7,8,9])
```

```python
from matplotlib import pyplot as plt
```

```python
def plot(r,t):
    beta = np.dot(r,t)/np.dot(r,r)
    t_par = r * beta
    t_orth = t - t_par
    plt.figure(figsize=(16,8))
    plt.plot([0,t[0]],[0,t[1]], label = r't')
    plt.plot([0,r[0]],[0,r[1]], label = r'r')
    plt.plot([0,t_par[0]],[0,t_par[1]], color = "k",linewidth = 3)
    plt.plot([0,t_orth[0]],[0,t_orth[1]],color=[.7,.7,.7], linewidth = 3)
    plt.show()
```

```python
t = np.random.randn(2)
r = np.random.randn(2)
plot(r,t)
```
