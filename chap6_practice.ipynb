```python
#exercise 6-2
def frob(v1,v2):
    return np.linalg.norm([v1,v2])
```

```python
N = 7
s = 1
nIter = 0
a = np.random.randn(N, N)
b = np.random.randn(N, N)
frob(a,b)
```

```python
while frob(a,b) > 1:
    s = 0.9 * s
    a = s * a
    b = s * b
    nIter += 1
print(nIter, s)
```
