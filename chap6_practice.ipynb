```python
#exercise 6-2
def frob(v1,v2):
    return np.linalg.norm([v1,v2])
```

```python
N = 7
s = 1
nIter = 0
a = np.random.randn(N, N)
b = np.random.randn(N, N)
frob(a,b)
```

```python
while frob(a,b) > 1:
    s = 0.9 * s
    a = s * a
    b = s * b
    nIter += 1
print(nIter, s)
```

```python
def rank_test(matrix,vector):
    try:
        Aug_m1 = np.concatenate((vector,matrix), axis=1)
        if np.linalg.matrix_rank(matrix) == np.linalg.matrix_rank(Aug_m1):
            return True
        else:
            return False
    except:
            print(f"Error: sizes don't match for augmentation!{matrix.shape},{vector.shape}")
    
```
